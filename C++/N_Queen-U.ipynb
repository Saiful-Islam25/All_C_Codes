```python
def is_safe(board, row, col, n):
    # Check column
    for i in range(row):
        if board[i][col] == 'Q':
            return False

    # left diagonal
    i, j = row - 1, col - 1
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j -= 1

    # right diagonal
    i, j = row - 1, col + 1
    while i >= 0 and j < n:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j += 1

    return True

def n_queens(board, row, n, ans):
    if row == n:
        ans.append(board[:])
        return

    for col in range(n):
        if is_safe(board, row, col, n):

            row_list = list(board[row])
            row_list[col] = 'Q'
            board[row] = ''.join(row_list)

            n_queens(board, row + 1, n, ans)

            row_list[col] = '*'
            board[row] = ''.join(row_list)


def solve_n_queens(n):
    board = ['*' * n for _ in range(n)]
    ans = []
    n_queens(board, 0, n, ans)
    return ans

solutions = solve_n_queens(4)
count = 0
for sol in solutions:
    count += 1
    print(f"\nSolution {count}:")
    for row in sol:
        print(row)

print(f"\nTotal solving ways: {count}")
```

```python
# একটি কুইন নির্দিষ্ট স্থানে বসানো নিরাপদ কিনা তা যাচাই করে
def is_safe(board, row, col, n):
    for i in range(row):
        if board[i][col] == 'Q':
            return False

    i, j = row - 1, col - 1
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j -= 1

    i, j = row - 1, col + 1
    while i >= 0 and j < n:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j += 1

    return True

# প্রথম সমাধান পাওয়ার পর থেমে যাবে
def n_queens_first_solution(board, row, n, ans):
    if row == n:
        ans.append(board[:])
        return True  # প্রথম সমাধান পেলে থেমে যাও

    for col in range(n):
        if is_safe(board, row, col, n):
            row_list = list(board[row])
            row_list[col] = 'Q'
            board[row] = ''.join(row_list)

            if n_queens_first_solution(board, row + 1, n, ans):
                return True  # সমাধান পেলে থেমে যাও

            row_list[col] = '*'
            board[row] = ''.join(row_list)

    return False

# শুধুমাত্র প্রথম সমাধান বের করার জন্য
def solve_first_n_queen(n):
    board = ['*' * n for _ in range(n)]
    ans = []
    n_queens_first_solution(board, 0, n, ans)
    return ans

# প্রথম সমাধান প্রিন্ট করি
solution = solve_first_n_queen(4)

if solution:
    print("Solution 1:")
    for row in solution[0]:
        print(row)
else:
    print("No solution found.")
```

```python
# একটি কুইন নির্দিষ্ট স্থানে বসানো নিরাপদ কিনা তা যাচাই করে
def is_safe(board, row, col, n):
    for i in range(row):
        if board[i][col] == 'Q':
            return False

    i, j = row - 1, col - 1
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j -= 1

    i, j = row - 1, col + 1
    while i >= 0 and j < n:
        if board[i][j] == 'Q':
            return False
        i -= 1
        j += 1

    return True

# সব সমাধান বের করি
def n_queens(board, row, n, ans):
    if row == n:
        ans.append(board[:])
        return

    for col in range(n):
        if is_safe(board, row, col, n):
            row_list = list(board[row])
            row_list[col] = 'Q'
            board[row] = ''.join(row_list)

            n_queens(board, row + 1, n, ans)

            row_list[col] = '*'
            board[row] = ''.join(row_list)

# দ্বিতীয় সমাধান বের করার জন্য
def solve_second_n_queen(n):
    board = ['*' * n for _ in range(n)]
    ans = []
    n_queens(board, 0, n, ans)

    # যদি কমপক্ষে দুইটি সমাধান পাওয়া যায়
    if len(ans) >= 2:
        return ans[1]  # দ্বিতীয় সমাধান রিটার্ন
    else:
        return None

# দ্বিতীয় সমাধান প্রিন্ট করি
solution = solve_second_n_queen(4)

if solution:
    print("Solution 2:")
    for row in solution:
        print(row)
else:
    print("Solution 2 not found.")
```
